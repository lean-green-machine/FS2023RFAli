```python
import cv2
import numpy as np

image = cv2.imread("cropfull100.jpg")#Reads in image file
cv2.waitKey(0) #Awaits user to press key
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #Converts image to grayscale
indices = np.where(gray <= 120) #Returns 2 arrays of y and x coordinates of defined line

#scale y by opening scale image, reading distance from top number to top of image
#then use ML to identify number s.t. we can create scale to multiple y axis values by

# key for Sorting tuple list by x axis
def xnum(n):  
    return n[0]    
     
# function to sort the coordinates    
def sort(list_of_tuples):  
    return sorted(list_of_tuples, key = xnum)    

coordinates = list((zip((4000 - indices[1]*(100/63)),(286 - indices[0])))) #zips up x array and y array to make coordinate list
coordinates = sort(coordinates) #sort coordinates by x axis

visited = set()
 
#Remove duplicates(takes top value of each x)
Output = []
for a, b in coordinates:
    if not a in visited:
        visited.add(a)
        Output.append((a, b))
        
coordinates = Output

cv2.waitKey(0) #necessary wait key


#write tuples into text file
f = open('usable_data_example.txt', 'w')
for t in coordinates:
    line = ' '.join(str(x) for x in t)
    f.write(line + '\n')
f.close()

#setup graph
import matplotlib.pyplot as plt
x_val = [x[0] for x in coordinates]
y_val = [x[1] for x in coordinates]

print (coordinates)
plt.plot(x_val,y_val)
plt.show()
```
